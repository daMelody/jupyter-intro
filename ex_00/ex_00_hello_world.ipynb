```python
import matplotlib.pyplot as plt
x=[5226,2404,3218,4321]
snacks=['BigMac','McSpicy','Cheeseburger','McNuggets']
colors=['r','m','y','g']
plt.axis('equal')
plt.pie(x,labels=snacks,colors=colors,shadow=True,explode=  
         (0.1,0,0,0),autopct=('%1.1f%%'))
plt.title('Snacks sold in a month')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pyp
from scipy import stats
x=np.array([1,2,3,4,5,6])
y=np.array([134,126,141,145,139,142])
a,b,correlation,p,error=stats.linregress(x,y)
print('Regression line: y=%.2fx+%.2f'% (a,b))
print('Correlation Coefficient: r=%.2f'% correlation)
plt.plot(x,y,'o',label='Original data')
f=a*x+b
plt.plot(x,f,'r',label='Regression Line')
plt.ylim(110, 160)
plt.legend()
plt.show()
```
